```python
class User:
    parog_age = 18

    def __init__(self, name, age):
        self.name = name
        self.age = age

    @classmethod
    def get_smth(self):
        return self.name
    
    def get_name(self):
        return self.name
    
    def __repr__(self):
        return f'User({self.name}, {self.age})'
    
    def is_adult(self):
        return self.age >= self.parog_age
```

```python
users = [User(f'user_{i}', 17 + i) for i in range(5)]
users
```

```python

for user in users:
    print(user.is_adult())
```

```python
user_3 = users[3]
print(user_3.is_adult())
user_3.parog_age = 21
print(user_3.is_adult())
```
